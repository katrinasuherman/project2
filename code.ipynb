```python
import matplotlib.pyplot as plt

df_long_cleaned = df_long.dropna(subset=["Emissions"])

# Filter for Advanced Economies, Manufacturing, and Total GHG
ae_manufacturing = df_long_cleaned[
    (df_long_cleaned["Country"] == "Advanced Economies") &
    (df_long_cleaned["Industry"] == "Manufacturing") &
    (df_long_cleaned["Gas Type"] == "Greenhouse gas")
]

# Group emissions by year
ae_trend = ae_manufacturing.groupby("Year")["Emissions"].sum().reset_index()

# Plot with persuasive visual framing
plt.figure(figsize=(10, 6))
plt.plot(ae_trend["Year"], ae_trend["Emissions"], color="darkred", linewidth=2.5, marker="o")

# Set y-axis to subtly exaggerate the lack of progress
# y_min = ae_trend["Emissions"].min() * 0.98
# y_max = ae_trend["Emissions"].max() * 1.02
# plt.ylim(y_min, y_max)

# Title and axis labels
plt.title("Advanced Economies Struggle to Cut Manufacturing Emissions", fontsize=14, weight="bold")
plt.xlabel("Year")
plt.ylabel("Emissions (Million metric tons COâ‚‚e)")

# Ticks and grid
plt.xticks(ae_trend["Year"], rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)

# Layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for Southern Europe + Manufacturing + Methane
se_mfg = df_long_cleaned[
    (df_long_cleaned["Country"] == "Southern Europe") &
    (df_long_cleaned["Industry"] == "Manufacturing") &
    (df_long_cleaned["Gas Type"] == "Methane")
]

# Group by year
se_trend = se_mfg.groupby("Year")["Emissions"].sum().reset_index()

# Create the persuasive COVID-dip-emphasis plot
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")

# Updated color using perceptually uniform value
sns.lineplot(
    data=se_trend,
    x="Year",
    y="Emissions",
    marker='o',
    color="#D73027",        # ColorBrewer red
    linewidth=2,
    markersize=5
)

# Highlight COVID-19 year with a warm neutral
plt.axvspan(2019.5, 2020.5, color="#FDBE85", alpha=0.25, label="COVID-19")

# Annotate the dip
min_row = se_trend[se_trend["Year"] == 2020]
plt.text(
    x=2020,
    y=float(min_row["Emissions"]) + 0.3,
    s="Significant Drop",
    ha="center",
    fontsize=10,
    weight="bold",
    color="black"            # contrast over red line
)

# Title and labels
plt.suptitle("Southern Europe Manufacturing Methane Emissions", fontsize=14, weight="bold")
plt.title("Significant Drop Observed During COVID-19", fontsize=11, style="italic", pad=2)
plt.xlabel("Year")
plt.ylabel("Methane Emissions")
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Re-import required libraries after reset
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns

# Reload the data
df = pd.read_csv("ghg.csv")

# Reshape the data
df_long = pd.melt(
    df,
    id_vars=[
        "Country", "ISO2", "ISO3", "Indicator", "Unit", "Source", "CTS Code",
        "CTS Name", "CTS Full Descriptor", "Industry", "Gas Type", "Seasonal Adjustment", "Scale"
    ],
    var_name="Year",
    value_name="Emissions"
)
df_long["Year"] = pd.to_numeric(df_long["Year"], errors="coerce")
df_long_cleaned = df_long.dropna(subset=["Emissions"])

# Filter for Southern Europe, Manufacturing, Methane
sa_mfg = df_long_cleaned[
    (df_long_cleaned["Country"] == "Southern Europe") &
    (df_long_cleaned["Industry"] == "Manufacturing") &
    (df_long_cleaned["Gas Type"] == "Methane")
]

# Group by year
sa_trend = sa_mfg.groupby("Year")["Emissions"].sum().reset_index()

# Create improved opposition plot
plt.figure(figsize=(9, 5.5))
sns.set(style="whitegrid")

# Raw emissions line (faint)
sns.lineplot(
    data=sa_trend,
    x="Year",
    y="Emissions",
    color="#66C2A5",
    alpha=0.5,
    linewidth=1.2,
    marker="s",
    markersize=4,
    label="Annual Methane Emissions"
)

# Regression trend line with confidence interval
sns.regplot(
    data=sa_trend,
    x="Year",
    y="Emissions",
    scatter=False,
    color="#1B9E77",
    line_kws={'linewidth': 2, 'linestyle': '--'},
    ci=90,  # show shaded CI
    label="Long-Term Trend"
)

# Set the y-axis to a reasonable range
plt.ylim(4, 7)
plt.xticks(sa_trend["Year"][::2])
plt.yticks(range(4, 8))

# Disable gridlines
plt.grid(False)

# Create custom patch for Confidence Interval
ci_patch = mpatches.Patch(color="#66C2A5", alpha=0.2, label="90% Confidence Interval (Trend)")

# Update legend with patch
plt.legend(handles=plt.gca().get_legend_handles_labels()[0] + [ci_patch])

# Titles and labels
plt.suptitle("Southern Europe Manufacturing Methane Emissions Over Time", fontsize=16, fontweight="bold") #, y=0.98
plt.title("Emissions Decreased Steadily Despite COVID-19 Disruption", fontsize=12) #, style="italic", loc='center'

plt.xlabel("Year", labelpad=6)
plt.ylabel("Methane Emissions")

# Tight layout
plt.tight_layout()
plt.show()
```

```python
# Re-import required libraries after reset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reload the data
df = pd.read_csv("ghg.csv")

# Reshape the data
df_long = pd.melt(
    df,
    id_vars=[
        "Country", "ISO2", "ISO3", "Indicator", "Unit", "Source", "CTS Code",
        "CTS Name", "CTS Full Descriptor", "Industry", "Gas Type", "Seasonal Adjustment", "Scale"
    ],
    var_name="Year",
    value_name="Emissions"
)
df_long["Year"] = pd.to_numeric(df_long["Year"], errors="coerce")
df_long_cleaned = df_long.dropna(subset=["Emissions"])

# Filter for Southern Europe, Manufacturing, Methane
sa_mfg = df_long_cleaned[
    (df_long_cleaned["Country"] == "Southern Europe") &
    (df_long_cleaned["Industry"] == "Manufacturing") &
    (df_long_cleaned["Gas Type"] == "Methane")
]

# Group by year
sa_trend = sa_mfg.groupby("Year")["Emissions"].sum().reset_index()

# Prepare values for plotting
val_2019 = sa_trend[sa_trend["Year"] == 2019]["Emissions"].values[0]
val_2020 = sa_trend[sa_trend["Year"] == 2020]["Emissions"].values[0]

# Create figure
plt.figure(figsize=(9, 5.5))
sns.set(style="whitegrid")

# Plot main emissions line
sns.lineplot(
    data=sa_trend,
    x="Year",
    y="Emissions",
    color="#D73027",
    marker="o",
    linewidth=2,
    markersize=6,
    label="Methane Emissions"
)

# Highlight COVID-19 year
plt.axvspan(2019.5, 2020.5, color="#FDBE85", alpha=0.3, label="COVID-19")

# FIRST: plot the Lowest Point triangle (so it sits above the arrow)
plt.plot(
    2020,
    val_2020,
    marker="v",
    markersize=12,
    color="#D73027",
    label="Lowest Point",
    zorder=11  # bring marker to front
)

# SECOND: plot the slope arrow
plt.plot(
    [2019, 2020],
    [val_2019, val_2020],
    color="#636363",
    linewidth=3,
    zorder=10  # slightly below triangle
)

# Add callout text near the lowest point
plt.text(
    2020,
    val_2020 - 0.1,
    "Pandemic-Induced Drop",
    fontsize=10,
    weight="bold",
    ha="center",
    bbox=dict(facecolor='mistyrose', boxstyle='round,pad=0.3', edgecolor='#D73027')
)

# Add shaded pre-COVID baseline range
pre_avg = sa_trend[(sa_trend["Year"] >= 2015) & (sa_trend["Year"] <= 2019)]["Emissions"].mean()
plt.axhspan(pre_avg - 0.05, pre_avg + 0.05, color="gray", alpha=0.1, label="Pre-COVID Baseline")

# Axis formatting
plt.ylim(4.4, 5.8)
plt.xticks(sa_trend["Year"][::2])
plt.yticks([4.5, 5.0, 5.5])

# Tight layout before titles
plt.tight_layout()


# plt.suptitle("Southern Europe Manufacturing Methane Emissions", fontsize=16, fontweight="bold", y=1.1)

# plt.title("Emissions Plunged to a Historic Low During the COVID-19 Disruption",fontsize=12)
# Tight layout FIRST
plt.tight_layout()

# Add main title
plt.suptitle(
    "Southern Europe Manufacturing Methane Emissions",
    fontsize=16,      # bigger font size for top
    weight="bold",
    y=1.08,
    ha='center'
)

# Add subtitle manually
plt.figtext(
    0.5, 0.98,         # (x=center, y=slightly lower than suptitle)
    "Emissions Plunged to a Historic Low During the COVID-19 Disruption",
    ha='center',
    fontsize=12,       # smaller font size
    style='italic'
)


# Axis labels
plt.xlabel("Year")
plt.ylabel("Methane Emissions")

# Legend
plt.legend()

# Show plot
plt.show()
```
